```python
import random
```

```python
class MarkovTextGenerator:
    def __init__(self):
        self.model = {}

    def train(self, text):
        words = text.split()

        for i in range(len(words) - 1):
            current_word = words[i]
            next_word = words[i + 1]

            if current_word not in self.model:
                self.model[current_word] = []

            self.model[current_word].append(next_word)

    def generate(self, start_word, length=20):
        if start_word not in self.model:
            return "Start word not found in training data."

        current_word = start_word
        result = [current_word]

        for _ in range(length - 1):
            next_words = self.model.get(current_word)
            if not next_words:
                break

            current_word = random.choice(next_words)
            result.append(current_word)

        return " ".join(result)
```

```python
text = """
machine learning is fun
machine learning is powerful
machine learning is everywhere
artificial intelligence is the future
machine intelligence is growing fast
"""
```

```python
generator = MarkovTextGenerator()
generator.train(text)

print("Model trained successfully!")
```

```python
print(generator.generate("machine", 15))
```

```python
print(generator.generate("artificial", 12))
```
